# Metodos de la clase Str
## Strip

```python
nombre = "                      angel ponce"
print(f"{nombre}")
nombre = nombre.strip()
print(f"{nombre}")
```

## Capitalize

```python
nombre = nombre.capitalize()
print(nombre)
```

## Title

```python
nombre = nombre.title()
print(nombre)
```

## Encadenar Metodos

```python
titulo = "      este es un titulo"
print(f"{titulo}")
titulo = titulo.strip().title()
print(titulo)
```

# Cambiar el case de la oracion

```python
titulo = titulo.swapcase()
print(titulo)
titulo = titulo.swapcase()
print(titulo)
```

# Dejar en mayusculas o en minusculas todo

```python
titulo = titulo.upper()
print(titulo)
titulo= titulo.lower()
print(titulo)
```

# Splitting Strings

```python
nombreCompleto = "Angel Chavez Ponce"
primero, segundo, tercero = nombreCompleto.split(" ")
print(primero)
print(segundo)
print(tercero)
```
